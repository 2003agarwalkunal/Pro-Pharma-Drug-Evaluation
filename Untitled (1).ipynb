```python
# Ques 6 Most widely used drug

most_used_drug = df.loc[df['Reviews'].str.replace(' Reviews', '').astype(int).idxmax()]
most_used_drug['Drug']
```

```python
# Most widely used drug
#using thsi approach we can find for other drugs as well

import seaborn as sns
import matplotlib.pyplot as plt

df['reviews_count'] = df['Reviews'].str.extract('(\d+)').astype(float)

# for better understanding we need to plot the top 5 used drugs
top_5_used = df.groupby('Drug')['reviews_count'].sum().nlargest(5)
# top_5_used

plt.figure(figsize=(10, 6))
top_5_used.plot(kind='bar', color='salmon')
plt.title('Top 5 Most Widely Used Drugs')
plt.xlabel('Drug')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=90)
plt.show()
```

```python
# group by drug and check for both indications 
drug_groups = df.groupby('Drug')['Indication'].unique()

drugs_with_both = [drug for drug, indications in drug_groups.items() if 'On Label' in indications and 'Off Label' in indications]

drugs_with_both
# list of all drugs with both indications
```

```python
# Q4
#calculate the average efficacy and average satisfaction of the drug 
# which are both off label and on label but only from the off label usage

if drugs_with_both:
    off_label_df = df[(df['Drug'].isin(drugs_with_both)) & (df['Indication'] == 'Off Label')]
    
    average_efficacy = off_label_df['Effective'].mean()
    average_satisfaction = off_label_df['Satisfaction'].mean()
    
    print(average_efficacy)
    print(average_satisfaction)
```

```python
#Q2
# drug which is used to treat the maximum number of conditions

# _grouping the data with the drug column and counting the number of unique conditions for each drug

drug_counts = df.groupby('Drug')['Condition'].nunique()

# drug used to treat maximum number of conditions
max_cond_drug = drug_counts.idxmax()
print(max_cond_drug)

# Number of conditions it treats
max_cond_count = drug_counts.max()
print(max_cond_count)
```

```python
# Q3 most effective drug to treat fever

fever_drugs = df[df['Condition'].str.contains('Fever', case = False)]
most_effective_fever_drug = fever_drugs.loc[fever_drugs['Effective'].idxmax()]
most_effective_fever_drug['Drug']
```
